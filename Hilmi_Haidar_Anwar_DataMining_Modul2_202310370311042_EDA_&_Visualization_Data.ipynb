## **A. Exploratory Data Analysis (EDA)**

Data is a collection of information or raw facts that can be symbols, numbers, words, or images. However, raw data is difficult to understand without adequate analysis. Exploratory Data Analysis (EDA) is an exploratory method used to understand the contents of a dataset through statistical summaries and graphical visualizations. EDA aims to find patterns, trends, and relationships between variables. Thus, data can be evaluated in depth to provide useful insights before proceeding to further analysis or building predictive models.

Data visualization is an important step in EDA, which is the process of displaying data in the form of graphs or diagrams to make it easier to understand. Data visualization helps to reveal patterns hidden in raw numbers, allowing users to see trends and relationships more intuitively. Some types of graphs that are often used include histograms for data distribution, bar charts for category comparisons, and pie charts for data proportions. Libraries that facilitate data processing and data visualization include **Matplotlib, Seaborn**, and **Pandas**.

### Handle Outliers

An outlier is an observation in a data set that has a different pattern from the other observations in that data set. This means that there are values that are much smaller or much larger than the other data in the data set. The causes of outliers include:

*   Errors in the data input procedure
*   Errors in measurement or analysis
* Special circumstances
Outlier


So, why is it important to identify outliers?
Outliers can affect the mean of the data and make the analysis results inaccurate. In addition, we can check whether there are errors in the data that need to be corrected. Sometimes, outliers are important data that represent extreme situations (e.g., very high-calorie foods).

Here are some ways to detect outliers:

#### Boxplot

Boxplots are an effective visual method for identifying outliers. Outliers are often displayed as points outside the boxplot's “whiskers.”

Let's try displaying outliers in one of the DataFrame columns.

````python
import matplotlib.pyplot as plt

plt.boxplot(df['calories'])
plt.title('Boxplot of Calories')
plt.ylabel('Calories')
plt.show()
````

We can analyze the visualization results above as follows:
1. The middle box represents data between the first quartile (Q1) and the third quartile (Q3). The line inside the box is the median. The calories data shows that the majority of calorie values are between 250 and around 750.
2. The whiskers (horizontal lines) show the “normal” data range without outliers. The calorie data shows that the upper limit of normal data is around 1000.
3. Points above or below the whiskers are **outliers**. These points are outside the normal range (smaller than the lower bound or larger than the upper bound). In the visualization above, the outliers are above 1000. This means that there are foods with calories ranging from 1500 to 2500, which are much higher than the other data points.

#### Z-Score

Z-Score measures how far an observation is from the mean, in standard deviation units. In other words, z-score shows whether a value is:

- Above average (positive z-score).
- Below average (negative z-score).
- Equal to average (z-score = 0)

By using Z-Score, we can detect outliers in data. Data with z > 3 or z < -3 are often considered outliers. Z-Score transforms data into a distribution with a mean of 0 and a standard deviation of 1, making it suitable for machine learning algorithms.

````python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df['calories']))
outliers = np.where(z_scores > 3)

print("Outliers index:", outliers)
````

#### Interquartile Range (IQR)

Another way to detect outliers is by using the Interquartile Range (IQR). IQR focuses on data quartiles, which divide data into four equal parts

> IQR = Q3 - Q1
* Q1 (First Quartile): The value below which 25% of the data lies.
- Q3 (Third Quartile): The value below which 75% of the data lies.
- IQR: The distance between Q3 and Q1, indicating the main data range.

We can use IQR to detect outliers if the data has a non-normal distribution. IQR is effective for detecting data that is far outside the main range.


````python
Q1 = np.percentile(df['calories'], 25)
Q3 = np.percentile(df['calories'], 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = np.where((df['calories'] < lower_bound) | (df['calories'] > upper_bound))
print("Outliers index:", outliers)
````

## **B. Data Visualization**

There are many types of data visualization, from simple to complex. Each type of visualization is usually more appropriate for a specific purpose. Broadly speaking, there are four general purposes of visualization.

1. Comparison

If you want to compare a set of values with other values, to see which is greater, which is higher, which is increasing, etc. For time-related data, the visualization used is usually a line chart. Meanwhile, if the comparison is categorical, then a bar chart is more suitable.

2. Distribution

If you want to see the distribution/spread of data in a variable, the visualization commonly used is a histogram. You can see whether the data is concentrated in small values, weighted to the right, or symmetrical, or whether it is normally distributed. To see the distribution in two variables, a scatterplot can also be used.

3. Composition

Used to see the composition of a variable when broken down by a data dimension. The visualization commonly used is a stacked bar chart for categorical data, or a stacked line chart for time-related data.

4. Relationship

Used to see the relationship between one variable and another. For example, to check whether variable A increases as variable B increases, or whether variable B decreases as variable A increases, or whether there is no relationship between the two variables. The visualization commonly used for this purpose is a scatter plot.

### Bar Chart

````python
import matplotlib.pyplot as plt
import numpy as np


data_bar = [20, 35, 30, 35, 27]

plt.figure(figsize=(13, 9))
plt.subplot(2, 2, 1) # row 2, col 2, position 1
plt.bar(range(len(data_bar)), data_bar)
plt.title('Bar Chart')
````

### Multi Bar-Chart

````python
import matplotlib.pyplot as plt
import numpy as np

multi_data_bar = {'Bar1': [20, 35, 30, 35, 27],
                  'Bar2': [25, 32, 34, 20, 25]}

plt.figure(figsize=(13, 9)) # Determine the visualization size
plt.subplot(2, 2, 2) # row 2, col 2, position 2
bar_width = 0.35
index = np.arange(len(multi_data_bar['Bar1']))

plt.bar(index, multi_data_bar['Bar1'], bar_width, label='Bar1')
plt.bar(index + bar_width, multi_data_bar['Bar2'], bar_width, label='Bar2')
plt.title('Multi Bar Chart')
plt.legend()
plt.show()
````

### Line Chart

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]


plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Prime Numbers', color='green',
         marker='o', linestyle='--')
plt.plot(x, y2, label='Pattern Numbers', color='red',
         marker='s')
plt.title('Line Chart', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)


plt.show()
````
